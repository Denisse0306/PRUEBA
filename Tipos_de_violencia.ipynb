```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path="/content/drive/MyDrive/Endireh/Endireh - Hoja 1.csv"
df = pd.read_csv(path)
fig, ax = plt.subplots()
plt.title("Mujeres que han sufrido violencia laboral")
x = np.array(["Total", "Psicológica", "Física y/o sexual"])
y = np.array([943235, 611518, 734118])
plt.bar(x, y, color = "green")
for bar in ax.patches:
  ax.text(bar.get_x() + bar.get_width() / 2,
          bar.get_height() / 2 + bar.get_y(),
          round(bar.get_height()), ha = 'center',
          color = 'w', weight = 'bold', size = 10)
plt.show()
```
